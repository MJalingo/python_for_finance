**Task: Drop the SEC Filings column**

```python
# CODE HERE
data1=data1.drop('SEC Filings',axis=1)
data1
```

**Task: Set the Symbol column to the index in the financials dataset**

```python
# CODE HERE
data1 = data1.set_index('Symbol')
data2 = data2.set_index('Symbol')
data1
```

**Task: What are the 10 largest companies according to market cap?**

```python
# CODE HERE
Sorted = data1['Market Cap'].sort_values(ascending=False)
Sorted.head(10)
# data1['Market Cap'].nlargest(10)
```

Note: The difference between GOOG and GOOGL is, that GOOG shares have no voting rights

**Task: Drop GOOG**

```python
# CODE HERE
data1 = data1.drop("GOOG",axis=0)
```

**Task: How many companies have a dividend yield > 4% ?**

```python
# CODE HERE
highest = data1[data1['Dividend Yield']>4].sort_values('Dividend Yield')
len(highest)
```

**Task: What is the mean Earnings per Share for all companies with a market cap > 1e+11**

```python
# CODE HERE
mean = data1[data1['Market Cap']>100000000000]['Earnings/Share'].mean()
mean
```

**Task: How many companies have a positive earnings per shares ratio?**

```python
# CODE HERE
pos = data1[data1['Earnings/Share']>0]
len(pos)
```

**Task: Which company pays the highest dividend yield? What was its 52 week high?**

```python
# CODE HERE
paymax = data1['Dividend Yield'].max()
high_pay = data1[data1['Dividend Yield']==paymax]['52 Week High']
high_pay
data1.loc[data1['Dividend Yield'].idxmax()]
```

**Task: Return the company with the largest spread between 52 weeks high and low**

**Task: Return all companies whith a price between \\$ 50 and \\$ 100**

```python
# CODE HERE
# data1[(data1['Price']>49) & (data1['Price']<101)]
data1[data1['Price'].between(50,100)]
```

**Task: Is there a correlation between the market cap and the Dividend Yield?**
